```python
sns.heatmap(compositeur_lieu, annot=True, cmap='viridis')
plt.show()
# on joue souvent Ravel dans l'agglo liloisse, jamais dans les concerts données à l'étranger
# La lecture des tableaux croisés conduit à supposer que les variables ne sont pas indépendantes. 
#le compositeur joué peut être prédit en partie  par la combinaison des autres variables
# la variable compositeur est binaire, ne comportant que deux modalités Ravel ou Mahler,
#cela se prête à une analyse prédictivce appuyée sur une classification (arebre de décision par exemple)
# Nous allons cependant devoir retravailler et recoder les variables 
# les AD sont plus efficaces particulièrement quand on travaille avec Pyton quand les var. sont numériques
```

```python
# les effectifs de certaines modalités de la variable Lieu sont faibles
# On peut regrouper certaines, mais il faut bien sûr que l'opération est un sens au regard des données
# Remplacement de 'E' et 'F' par 'Hors', et conservation de 'A'
# Nous venons d'introduire nos premières instructions conditionelles (if et else). Nous les retrouverons

df['LIEU'] = df['LIEU'].apply(lambda x: 'Hors' if x in ['E', 'F'] else x)
```

```python
print(df.head())
#on vérifie (toujours)que le résultat est bien le résultat attendu
```

```python
#pOUR mes analyses il serait mieux de tranformer mes variables quali en variable quanti
# je veux créer deux nouvelles varaibles dans le dataframe obtenu en lisant la variable A
# je peux donc dans mon carnet créer de nouvelles variables en manipulant celles déjà présentes

df['Ag'] = [1 if x == 'A' else 0 for x in df['LIEU']]
df['Hag'] = [1 if x == 'Hors' else 0 for x in df['LIEU']]

```

```python
print(df.head())
#on vérifie (toujours)que le résultat est bien le résultat attendu
```

```python
# j'ai modifié mon fichier de données je voudrais garder une copie des données modifiées
# je souhaite le faire en les enregistrant dans un nouveau fichier xl 

df.to_excel("musicrecode.xlsx", index=False)
#je dois indiquer le chemin vers l'emplacement désiré et le nom du nouveau ficier
#en ce cas il est enregistré sous le nom musicrecode.xlsx dans le même répertoire que le carnet
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder


X = df[['ANNEE', 'Ag', 'Hag']]  # Variables explicatives
y = df['COMPOSITEUR']  # Variable à expliquer

# Encodage des labels car 'COMPOSITEUR' est une variable qualitative
label_encoder = LabelEncoder()
y_encoded = label_encoder.fit_transform(y)

# Division en ensemble d'entraînement et de test
X_train, X_test, y_train, y_test = train_test_split(X, y_encoded, test_size=0.3, random_state=42)
```

```python


from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.metrics import accuracy_score

# Entraînement du modèle
tree_model = DecisionTreeClassifier(random_state=42)
tree_model.fit(X_train, y_train)

# Conversion de label_encoder.classes_ en liste
class_names_list = label_encoder.classes_.tolist()

# Prédiction et évaluation du modèle
y_pred = tree_model.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy de l'Arbre de Décision: {accuracy}")

# Visualisation de l'arbre
plt.figure(figsize=(20,10))  
plot_tree(tree_model, filled=True, feature_names=['ANNEE', 'Ag', 'Hag'], class_names=class_names_list)
plt.show();
```

Le modèle n'est pas très efficace (il prédit le compositeur dans les deux tiers des cas, ce qui vu les effectifs de la population et la configuration des varialbes - trop de modalités pour chefs que nous ne pouvons donc pas prendre en compte - n'est pas si mal. 
Il met en évidence qu'au regard des données le caractère le plus fortement associé au nom du compositeur est le lieu du concert. L'orchestre tend à jouer Mahler quand il n'est pas à proximité de son siège, avec possiblement, mais les coefficients sont faibles, une incidence de l'année durant lequel se produit le concert.
