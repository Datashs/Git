# D'un tableau Excel illisible à un Json propre

Il arrive assez régulièrement que vous soyiez conraint de contrôler un fichier XL d'une taille déraisonnable, dont les colonnes ne tiendraient pas sur un écran géant, ce qui occasionne une gymnastique qui peut agacer. <br>
Une solution possible est de transformer cela en un fichier json, que l'on peut ensuite soit imprimer en pdf s'il ne s'agit que de le lire, soit ouvrir avec un éditeur de code si l'on veut inervenir sur les données. <bR>
Il existe des solutions gratuites en ligne, mais elles supposent généralement que l'on ouvre un compte, donne ses coordonnées de carte bancaire, pour ne pas parler de la confidentialilté des données. <br>
Il est tout simple avec Python de le faire soi même. Objet de ce petit carnet. 
Si jamais je trouve le temps je ferai peut être une petite application, plus pratique qu'un carnet. Mais bon, ce n'est pas certain non plus. 

```python
import pandas as pd

# Charger le fichier Excel
# Le fichier est ici dans le même dossier que le carnet
# Les sorties sont enregistrées dans le même dossier. Vous pouvez bien sûr changer le nom des fichiers en entrée et en sortie. 
df = pd.read_excel('cand.xlsx')

# Afficher les premières lignes pour vérifier le contenu
print(df.head())
# une routine simpole pour récupérer des fichiers lisibles. Prend une feuille xl en entrée, retourne du json, et enregistre le json
```

```python
# Convertir le DataFrame en JSON
json_data = df.to_json(orient='records', force_ascii=False)
# L'argument orient dans la méthode to_json() spécifie la manière dont les données sont formatées dans le JSON. 
#L'option 'records' indique que chaque ligne du DataFrame sera convertie en un objet JSON. 
#Chaque ligne devient donc un objet JSON où les noms de colonnes deviennent des clés 
# l'option force_ascii=False permet de conserver les caractères non ascii tels quels (caractères accentués, alphabets non latins

# Afficher les données JSON
print(json_data)
```

```python
import json

# Écrire le JSON dans un fichier
with open('donnees.json', 'w', encoding='utf-8') as f:
    json.dump(json.loads(json_data), f, ensure_ascii=False, indent=4)
#encoding='utf-8' spécifie que le fichier doit utiliser l'encodage UTF-8. 
#Cela permet un traitement correct des caractères non-ASCII.
#indent=4 : Cette option formatte le JSON afin qu'il soit lisible par les humains. 
#Elle ajoute un indentation de quatre espaces à chaque niveau de la structure du JSON

```
