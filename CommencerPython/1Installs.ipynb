### Modules 

Un module python est un fichier contenant un code qui permet de définir des variables, des fonctions. Il prend la forme d'un ficheir se terminant en .py. Un module peut être pensé comme une petite collection de fonctions utilisables dans un même contexte. </br>
Le module 'math' fournit ainsi un ensemble de fonctons mathématiques que l'on pourra appeler dans le cours d'un développement, s'il est installé dans l'environnement que l'on a crée pour accomplir une tâche.

### Bibliothèques

Une bibliothèque rassemble un ensemble de modules, de fonctions, de classes utilisées pour étendre les capacités de python dans un domaine particulier. Vous utiliserez probablement Pandas, dédiée à la manipulation de données et Matplotlib, spécialisée dans la visualisation de données.

### Environnements 

Un environnemnet est un espace de travail, isolé de vos autres instances de python, qui contient une version spécifrique de python et ensemble de bilbliothèques et de fonctions </br>
Cela permet de gérer les dépendances de manière indépendante pour différents projets, évitant ainsi les conflits entre les versions de bibliothèques ou de Python lui-même.

## Bonnes pratiques 

Il est fermement conseillé de définir un environnement spécifique pour chaque projet, voire pour chaque type de tâches (si par exemple vous voulez faire de la carthographie).
